```python
import matplotlib.pyplot as plt
from scipy import ndimage,misc
import cv2
import numpy as np
```

```python
img= cv2.imread("images/sultnois.jpg")
#روش اول
result1=ndimage.median_filter(img, size=7)
#روش دوم
result2= cv2.medianBlur(img,9)
plt.subplot(3,1,1)
plt.imshow(img)
plt.subplot(3,1,2)
plt.imshow(result1)
plt.subplot(3,1,3)
plt.imshow(result2)
```
