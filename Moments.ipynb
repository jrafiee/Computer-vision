
1 .گشتاورهاي ثابت نسبت به تغيير کنتراست

```python

#Test3-1 ,Test3-2,Test3-3
plt.figure(figsize=(15,10))
c=0.3;# Contrast Factor

img=cv2.imread('Pictures/12.png',0)
img1=c*img

m1=cv2.moments(img)
m2=cv2.moments(img1)
dist1=m1['m20']-m2['m20']
dist2=(m1['m20']/m1['m00'])-(m2['m20']/m2['m00'])


w1=plt.subplot(1,2,1)
plt.imshow(img,cmap='gray', vmin = 0, vmax = 255)
w1.title.set_text("input Image")
w2=plt.subplot(1,2,2)
plt.imshow(img1,cmap='gray', vmin = 0, vmax = 255)
w2.title.set_text("with c="+str(c)+"\n Spatial moment="+str(dist1)+"\n Spatial moment2="+str(dist2))
#مشاهده می شود که گشتاور هندسی نسبت به تغییر کنتراست ثابت نیست ولی با تغییر در  آن می توان آن را اصلاح کرد
```


2 .گشتاورهاي ثابت نسبت به انتقال

```python
plt.figure(figsize=(15,10))
img1=cv2.imread('Pictures/20.bmp',0)
img2=cv2.imread('Pictures/21.bmp',0)
m1=cv2.moments(img1)
m2=cv2.moments(img2)
dist1=m1['m20']-m2['m20']
dist2=(m1['m20']/m1['m00'])-(m2['m20']/m2['m00'])
dist3=m1['mu20']-m2['mu20']

x1=plt.subplot(1,2,1)
plt.imshow(img1,cmap='gray',vmin=0,vmax=255)
x2=plt.subplot(1,2,2)
plt.imshow(img2,cmap='gray',vmin=0,vmax=255)
x2.title.set_text(" Spatial moment="+str(dist1)+"\n Spatial moment2="+str(dist2)+"\n Centeral moment="+str(dist3))
#مشاهده می شود گشتاور هندسی نسبت به جا به جایی تصویر ثابت نیست 
#اما گشتاور مرکزی ثابت می ماند
```


3 .گشتاورهاي ثابت نسبت به تغيير مقياس

<br/>
بررسي ثابت بودن گشتاورمرکزي نرمال شده نسبت
به تغيير مقياس

```python
# Test9.m
plt.figure(figsize=(15,10))
img1=cv2.imread('Pictures/20.bmp',0)
S=2;# Scale Factor
x,y=img1.shape
print(x)
print(y)
img2=cv2.resize(img1,(S*y,S*x))
m1=cv2.moments(img1)
m2=cv2.moments(img2)
dist1=m1['m20']-m2['m20']
dist2=(m1['m20']/m1['m00'])-(m2['m20']/m2['m00'])
dist3=m1['mu20']-m2['mu20']
dist4=m1['nu20']-m2['nu20']


x1=plt.subplot(1,2,1)
plt.imshow(img1,cmap='gray',vmin=0,vmax=255)
x2=plt.subplot(1,2,2)
plt.imshow(img2,cmap='gray',vmin=0,vmax=255)
x2.title.set_text(" 'Spatial moment'="+str(dist1)+"\n Spatial moment2="+str(dist2)+"\n Centeral moment="+str(dist3)+"\n Normalized central moment="+str(dist4))
#مشاهده می شودکه گشتاور هندسی و گشتاور مرکزی نسبت به تغییر مقیاس ثابت نیستند 
#اما گشتاور مرکزی نرمال شده نسبت به تغیر مقیاس ثابت است
```
